```python
import spacy
nlp = spacy.load("en_core_web_md") # modelo de nlp
```

```python
dict_tipos_palabras=dict()
dict_tipos_palabras_dep=dict()
```

```python
for index, strings in rte.iterrows():
    text = str(strings["sentence1"]).lower()
    #print(text)
    # Process the document via Spacy's nlp
    doc = nlp(text)
    for token in doc:
        #print(token)
        # if token.pos_ in ["NUM","PROPN","NOUN","VERB","ADJ","ADV"] or token.dep_ =="neg":
        #     if token.lemma_ !="be":
        #         words.append(token.lemma_)
        # print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
        #     token.shape_, token.is_alpha, token.is_stop)
        if token.pos_ in dict_tipos_palabras:
            a=set(dict_tipos_palabras[token.pos_])
            a.add(token.lemma_)
            #print(a)
            dict_tipos_palabras[token.pos_]=a
        else:
            dict_tipos_palabras[token.pos_]=set([token.lemma_])
        if token.dep_ in dict_tipos_palabras_dep:
            a=set(dict_tipos_palabras_dep[token.dep_])
            a.add(token.lemma_)
            #print(a)
            dict_tipos_palabras_dep[token.dep_]=a
        else:
            dict_tipos_palabras_dep[token.dep_]=set([token.lemma_])
    text = str(strings["sentence2"]).lower()
    #print(text)
    # Process the document via Spacy's nlp
    doc = nlp(text)
    for token in doc:
        #print(token)
        # if token.pos_ in ["NUM","PROPN","NOUN","VERB","ADJ","ADV"] or token.dep_ =="neg":
        #     if token.lemma_ !="be":
        #         words.append(token.lemma_)
        # print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
        #     token.shape_, token.is_alpha, token.is_stop)
        if token.pos_ in dict_tipos_palabras:
            a=set(dict_tipos_palabras[token.pos_])
            a.add(token.lemma_)
            #print(a)
            dict_tipos_palabras[token.pos_]=a
        else:
            dict_tipos_palabras[token.pos_]=set([token.lemma_])
        if token.dep_ in dict_tipos_palabras_dep:
            a=set(dict_tipos_palabras_dep[token.dep_])
            a.add(token.lemma_)
            #print(a)
            dict_tipos_palabras_dep[token.dep_]=a
        else:
            dict_tipos_palabras_dep[token.dep_]=set([token.lemma_])
```
